```python
#Filter genres according the exercise
genre_filters = ['action','comedy','adventure','animation','drama']
data2 = data[data.Genre.isin(genre_filters)]
```

```python
#Define the style
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(9,6))

#Plot the boxplots
b = sns.boxplot(data=data3, x="Genre", y="Gross % US", color='lightgray', showfliers=False)
plt.setp(b.artists, alpha=0.5, ) #contorno boxplots

sns.stripplot(x='Genre', y='Gross % US', data=data3, size=6, hue='Studio', alpha=0.7)
plt.title("Domestic Gross % by Genre", fontsize=30)
plt.xlabel("Genre", fontsize=20)
plt.ylabel("Gross % US", fontsize=20)

plt.legend(bbox_to_anchor=(1.05,1), framealpha=0, loc="upper left")
#plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
#Importing the csv
data = pd.read_csv("P4-Section6-Homework-Dataset.csv", encoding='latin1')
```

```python
#Exploring the dataset
#data.info()
#data.describe()
data.head()
```

```python
#Changing column names
data.columns = ['DayOfWeek', 'Director', 'Genre', 'MovieTitle', 'ReleaseDate',
       'Studio', 'AdjustedGross', 'Budget', 'Gross',
       'IMDbRating', 'MovieLensRating', 'Overseas', 'Overseas%',
       'Profit', 'Profit%', 'Runtime', 'US',
       'Gross%US']
```

```python
#Changing into categories makes the charts consider all genres/studios, even with filtering
#data.DayOfWeek = data.DayOfWeek.astype('category')
#data.Director = data.Director.astype('category')
#data.Genre = data.Genre.astype('category')
#data.MovieTitle = data.MovieTitle.astype('category')
#data.ReleaseDate = data.ReleaseDate.astype('category')
#data.Studio = data.Studio.astype('category')
```

```python
#Filter genres
genres_filter = ['action', 'comedy', 'adventure', 'animation', 'drama']
data_genres = data[data.Genre.isin(genres_filter)]
```

```python
#Filter studios
studios_filter = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']
data_studios = data_genres[data_genres.Studio.isin(studios_filter)]
```

```python
data_studios.info()
```

```python
#Define the style
sns.set_style(style="darkgrid")
f, ax = plt.subplots(figsize=(9,6))

#Plot the boxplots
b = sns.boxplot(data=data_studios, x='Genre', y='Gross%US', color='lightgray', showfliers=False)
plt.setp(b.artists, alpha=0.5, ) #contorno boxplots

#Plot the stripplots
s = sns.stripplot(data=data_studios, x='Genre', y='Gross%US', hue='Studio', size=6, alpha=0.5)

plt.title('Domestic Gross % by Genre', fontsize=30)
plt.xlabel('Genre', fontsize=20)
plt.ylabel('Gross % US', fontsize=20)
plt.yticks(np.arange(0, 100, 20))
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left', framealpha=0)
plt.show()
```
