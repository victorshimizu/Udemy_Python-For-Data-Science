```python
#Read the csv
data = pd.read_csv('P4-Section6-Homework-Dataset.csv', encoding='latin1')
```

```python
vis1 = sns.factorplot(data=data, x='Day of Week', kind='count', size=6)
```

```python
#Filter genres according the exercise
genre_filters = ['action','comedy','adventure','animation','drama']
data2 = data[data.Genre.isin(genre_filters)]
```

```python
#Filter studios according the exercise
studios_filters = ['Buena Vista Studios','Sony','Universal','WB','Paramount Pictures','Fox']
data3 = data2[data2.Studio.isin(studios_filters)]
```

```python
#Define the style
sns.set(style="darkgrid")
fig, ax = plt.subplots(figsize=(9,6))

#Plot the boxplots
b = sns.boxplot(data=data3, x="Genre", y="Gross % US", color='lightgray', showfliers=False)
plt.setp(b.artists, alpha=0.5, ) #contorno boxplots

sns.stripplot(x='Genre', y='Gross % US', data=data3, size=6, hue='Studio', alpha=0.7)
plt.title("Domestic Gross % by Genre", fontsize=30)
plt.xlabel("Genre", fontsize=20)
plt.ylabel("Gross % US", fontsize=20)

plt.legend(bbox_to_anchor=(1.05,1), framealpha=0, loc="upper left")
#plt.show()
```
